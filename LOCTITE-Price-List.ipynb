```python
# List of tables
tables = [table_1, table_2, table_3, table_4, table_5,
          table_6, table_7, table_8, table_9, table_10,
          table_11, table_12, table_13, table_14, table_15]

# Create dictionary to store the information
dict1 = {'ID': [], 'Title': [], "Description": [], "Price": [], "Attributes": {"Pack Size": []}}

# Iterate through each table
for table in tables:
    # Iterate through each row in the table
    for row in table:
        # Extract information from the row
        sku_id, title, description, price, pack_size = row

        # Append the information to the dictionary
        dict1['ID'].append(sku_id)
        dict1['Title'].append(title)
        dict1['Description'].append(description)
        dict1['Price'].append(price)
        dict1['Attributes']['Pack Size'].append(pack_size)

# Print the dictionary for verification
print(dict1)
```
